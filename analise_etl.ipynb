# Extract

```python
import pandas as pd

df = pd.read_csv("Financials.csv")
display(df)
```

# Transform

```python

print(df.columns)
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
df = df.replace({'\$': '', '-': ''}, regex=True)
display(df)
```

```python
valor =' None '

df = df.drop(df[df['Discount Band'] == valor].index)
display(df)
```

# Load

```python
nome_arquivo_csv = "new_financials.csv"

df.to_csv(nome_arquivo_csv, index=False)
```
